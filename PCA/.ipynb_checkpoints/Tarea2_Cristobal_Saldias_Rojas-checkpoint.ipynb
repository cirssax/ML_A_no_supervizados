```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Previsualización de la información
Se muestran las columnas que componen el dataset facilitado de nombre: brandrating.csv para que posteriormente, se muestren estadígrafos básicos, cantidad de nulos por columna (para previsualizar si será necesario realizar imputación de valores) y finalmente, los tipos de variables a los que corresponde cada columna.

```python
data = pd.read_csv("brandrating.csv")
print("=================================================================================")
print("Columnas dataset")
print(data.columns)
print("=================================================================================")
print("Estadisticos numericos dataset")
print(data.describe())
print("=================================================================================")
print("Cantidad de nulos por columna")
print(data.isna().sum())
print("=================================================================================")
print("Tipo de variable de cada columna")
print(data.dtypes)
print("=================================================================================")
print("Tamaño dataset")
print(data.shape)
print("=================================================================================")
print("Dataset")
data.head()
```

```python
# Se copian las variables numércias a un data set nuevo
data_num = data[["perform","leader","latest","fun","serious","bargain", "value", "trendy", "rebuy"]]
data_num.head()
```

## Mapa de calor por correlacion de variables del dataset

```python
Var_Corr = data_num.corr()
plt.figure(figsize = (12,10))
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
plt.show()
```

## Componentes principales

```python
from sklearn.preprocessing import StandardScaler #estandarizar las variables
from numpy.linalg import eig #eginvalues
import numpy.linalg as la

# Se copian las columnas numericas a un nuevo dataset
data_numerico = data_num
scaler = StandardScaler()
data_standarizada = scaler.fit_transform(data_numerico)# Estandarizacion de variables
print(data_standarizada)
```

```python
# Se revisa si la estandarizacion esta correcta
promedio = np.round(np.mean(data_standarizada), decimals = 2)
desv_std = np.round(np.std(data_standarizada), decimals = 2)
print("Promedio: ", promedio)
print("Desviacion estandar:", desv_std)
```

```python
# Se transforma a un dataframe los campos estandarizados
data_std = pd.DataFrame(data_standarizada, columns = ["perform","leader", "latest", "fun", "serious", "bargain", "value", "trendy", "rebuy"])
data_std.head()
```

```python
# Se calcula la covarianza
data_cov = np.cov(data_std)
print("================================================================")
print("Matriz de Covarianza de dataset normalizado")
print(data_cov)
```

```python
# Se obtienen los eigenvalues y eigenvectors
eigvalues, eigvectors = eig(data_cov)
print("=================================================================")
print("Cantidad de eigvectors generados:")
print(eigvectors.shape)
```

```python
# Calculo de componentes principales
from sklearn.decomposition import PCA
pca = PCA(n_components=6) #cuantas dimensiones de PCA
pca.fit(data_standarizada)
```

```python
# Se convierten los datos a las nuevas dimensiones
pca_n = pca.transform(data_standarizada)
print(pca_n)
print(pca_n.shape)
```

## Varianza explicada hasta la 3° componente

```python
print('Varianza explicada PCA: ', pca_n.shape)
expl = pca.explained_variance_ratio_
print("=================================================================")
print("% de datos faltantes por cada componente: ")
for i, elem in enumerate(expl):
    print((np.round(elem, decimals = 4) * 100), "% faltante para la componente principal n°", i)
print("=================================================================")
print("Porción de la varianza explicada acumulada hasta la 3° componente: ", (np.round(sum(expl[0:3]), decimals = 4)))
```

## Gráfico de la varianza explicada

```python
# graficar la curva de varianza acumulada
plt.figure(figsize=(15,8))
plt.grid()
plt.title("Varianza explicada", fontsize = 20)
plt.xlabel("Nº de componentes", fontsize=13)
plt.ylabel("varianza explicada acumulada", fontsize=13)

plt.yticks(fontsize = 13)
plt.xticks(fontsize = 13)

plt.plot(np.cumsum(pca.explained_variance_ratio_), marker="*", linewidth=4.0, alpha = 0.6, c = "#FFC107")

for index,elem in enumerate(np.cumsum(pca.explained_variance_ratio_)):
    value = np.round(elem, decimals = 4)
    plt.text(x = index, y = elem, s = f"{value}")

plt.show()
```

## Dataset y mapa de calor de componentes principales

```python
data_comp = pd.DataFrame(pca.components_, columns=["perform","leader", "latest", "fun", "serious", "bargain", "value", "trendy", "rebuy"])
data_comp
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(data_comp, cmap='plasma', annot = True)
plt.show()
```

## Relación "componente principal 2" respecto a variable "leader"
La componente principal 2 explica en gran medida la variable leader, para ser exactos la explica con un valor de 0.52. Esto quiere decir que la componente 2 explica en gran medida la variable leader, por lo que si esa variable es una que se desea someter a estudio más profundo, sería recomendable emplear la segunda componente.
